```python
import os
from pydub import AudioSegment
from tqdm import tqdm
import random

folder_path = r"./kikyo_wav\kikyo"
output_file = AudioSegment.empty()
output_files = []
max_length = 15 * 1000
pause_length = 100
l = os.listdir(folder_path)
random.shuffle(l)
for filename in l:
    file_path = os.path.join(folder_path, filename)
    if file_path.endswith(".wav"):
        audio = AudioSegment.from_file(file_path, format="wav")
        if len(audio) < 5000:
            if len(output_file) + len(audio) + pause_length <= max_length:
                output_file += AudioSegment.silent(duration=pause_length) + audio
            else:
                output_files.append(output_file)
                output_file = audio


if len(output_file) > 0:
    output_files.append(output_file)

for i, output_file in tqdm(enumerate(output_files)):
    output_file.export(folder_path+f"/concentrate_1_{i}.wav", format="wav")
output_file
```

```python
import os

def rename_files(folder_path):
    for filename in os.listdir(folder_path):
        if filename.startswith('concentrate'):
            if filename.endswith(".wav"):
                file_path = os.path.join(folder_path, filename)
                name, extension = os.path.splitext(filename)
                parts = name.split("_")
                if len(parts) >= 3 and parts[1].isdigit():
                    number = int(parts[1])
                    new_number = number + 10
                    new_name = f"{parts[0]}_{new_number}_{parts[2]}{extension}"
                    new_file_path = os.path.join(folder_path, new_name)
                    os.rename(file_path, new_file_path)


folder_path = r'D:\Code\Python396\Wakamo'
rename_files(folder_path)
```
